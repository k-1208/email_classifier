```python
# Normalize labels to spam/valid/marketing in full_dataset.cleaned.jsonl
import json
from pathlib import Path

INPUT_JSONL = Path("/home/kunal/jhinga/fine-tune/fine-tune-email/dataset_part2/cleaned/full_dataset.cleaned.jsonl")
OUTPUT_JSONL = Path("/home/kunal/jhinga/fine-tune/fine-tune-email/dataset_part2/cleaned/full_dataset.cleaned.labels.jsonl")

def map_label(value: str | None) -> str:
    if value is None:
        return "valid"
    val = str(value).strip().lower()
    if val == "spam":
        return "spam"
    if val == "promotions":
        return "marketing"
    if val == "marketing":
        return "marketing"
    return "valid"

count = 0
with INPUT_JSONL.open("r", encoding="utf-8") as f_in, OUTPUT_JSONL.open("w", encoding="utf-8") as f_out:
    for line in f_in:
        if not line.strip():
            continue
        obj = json.loads(line)
        obj["label"] = map_label(obj.get("category"))
        obj.pop("category", None)
        f_out.write(json.dumps(obj, ensure_ascii=False) + "\n")
        count += 1

print(f"Wrote {count} records to {OUTPUT_JSONL}")
```

```python
# Map label 1 -> phishing in Nazario.cleaned.dedup.jsonl
import json
from pathlib import Path

INPUT_JSONL = Path("/home/kunal/jhinga/fine-tune/fine-tune-email/dataset_part2/cleaned/Nazario.cleaned.dedup.jsonl")
OUTPUT_JSONL = Path("/home/kunal/jhinga/fine-tune/fine-tune-email/dataset_part2/cleaned/Nazario.labeled.jsonl")

def map_label(value) -> str:
    if value is None:
        return "valid"
    # Handle int-like strings too
    try:
        numeric = int(value)
    except (TypeError, ValueError):
        return str(value)
    return "phishing" if numeric == 1 else "valid"

count = 0
with INPUT_JSONL.open("r", encoding="utf-8") as f_in, OUTPUT_JSONL.open("w", encoding="utf-8") as f_out:
    for line in f_in:
        if not line.strip():
            continue
        obj = json.loads(line)
        obj["label"] = map_label(obj.get("label"))
        f_out.write(json.dumps(obj, ensure_ascii=False) + "\n")
        count += 1

print(f"Wrote {count} records to {OUTPUT_JSONL}")
```
